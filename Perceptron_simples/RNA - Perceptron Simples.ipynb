```python
import pandas as pd
import numpy as np
```

```python
class percpetron:
    def __init__(self, amostra, dy):
        self.pesoW0 = np.random.rand(1,1)
        self.pesoW1 = np.random.rand(1,1)
        self.pesoW2 = np.random.rand(1,1)
        self.pesoW3 = np.random.rand(1,1)
        self.epocas = 0
        self.eta = 0.01
        self.entrada = amostra
        self.dy = dy
        self.erro = True
        self.bias = -1
    
    def treino(self):
        maiorAcerto = 0
        while(self.erro):
            acertou = 0
            self.epocas = self.epocas + 1
            self.erro = False
            for row in range(self.entrada.shape[0]):
                x = self.entrada.iloc[row].to_numpy()
                y = self.saida(x[0],x[1],x[2])
                dy = self.dy[row]
                if(y != dy):
                    self.pesoW0 = self.pesoW0 + (self.eta * (dy - y) * self.bias)
                    self.pesoW1 = self.pesoW1 + (self.eta * (dy - y) * x[0])
                    self.pesoW2 = self.pesoW2 + (self.eta * (dy - y) * x[1])
                    self.pesoW3 = self.pesoW3 + (self.eta * (dy - y) * x[2])
                    self.erro = True
                else:
                    acertou = acertou + 1
            if (acertou > maiorAcerto):
                maiorAcerto = acertou 
            if (self.epocas > 5000):
                print(f"Rede não convergiu em {self.epocas }")
                print(f"maior acerto {maiorAcerto}")
                break
        print(f"Rede aprendeu em {self.epocas} épocas")
                
    def addBias(self):
        bias_array = np.ones(amostra.shape[0])
        bias_array = np.dot(bias_array, self.bias)
        self.entrada.insert(0,'x0',bias_array, False)
        
    def funcaoAtivacao(self, u):
        if u >= 0:
            return 1
        else:
            return -1

    def saida(self, x1, x2, x3):
        u = self.somaPonderada(x1, x2, x3) 
        return self.funcaoAtivacao(u)
    
    def somaPonderada(self, x1, x2, x3):
        u = ( (self.pesoW1 * x1) + (self.pesoW2 * x2) + (self.pesoW3 * x3) + (self.bias * self.pesoW0) )    
        return u
    
    def classificar(self, amostra):
        result = {}
        for row in range(amostra.shape[0]):
            x = self.entrada.iloc[row].to_numpy()
            y = self.saida(x[0],x[1],x[2])
            if (y == -1):
                result[f'amostra_{row}'] = 'P1'
            else:
                result[f'amostra_{row}'] = 'P2'
                
            
        return result
```

```python
df = pd.read_csv('files/treinamento.csv', sep=';')
treino = df[['x1', 'x2', 'x3']]
dy = df['d']
```

```python
amostra = pd.read_csv('files/amostra.csv', sep=';')
amostra = amostra[['x1', 'x2', 'x3']]
```

```python
perceptron = percpetron(treino, dy)
print("======== 1 ========")
print(f"""Pesos iniciados aleatóriamente: 
{perceptron.pesoW0}\n{perceptron.pesoW1}\n{perceptron.pesoW2}\n{perceptron.pesoW3}""")
print("*"*50)
perceptron.treino()
print(f"""Pesos ajustados: 
{perceptron.pesoW0}\n{perceptron.pesoW1}\n{perceptron.pesoW2}\n{perceptron.pesoW3}""")
print("*"*50)
print(f"Classificação da amostra")
resultado = perceptron.classificar(amostra)
resultado = pd.DataFrame.from_dict(resultado, orient='index', columns=['Classificação'])
resultado
```
